Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Data Collection and Processing

```python
#loading the data from csv file to a pandas dataframe
calories=pd.read_csv('/content/calories.csv')
```

```python
#print the first 5 rows of the dataframe
calories.head()
```

```python
exercise_data=pd.read_csv('/content/exercise.csv')
```

```python
exercise_data.head()
```

```python
calories_data=pd.concat([exercise_data, calories['Calories']],axis=1)
```

```python
calories_data.head()
```

```python
#finding the rows and columns
calories_data.shape
```

```python
#printing the rows and columns seperately
print("Number of rows:", calories_data.shape[0])
print("Number of columns:", calories_data.shape[1])
```

```python
#getting some information about the data
calories_data.info()
```

```python
#checks for missing values
calories_data.isnull().sum()
```

Data Analysis

```python
#get some statistical measures about the data
calories_data.describe()
```

Data Visualisation

```python
#plotting the gender column in a count plot
sns.countplot(x='Gender', data=calories_data)
```

```python
#finding the distribution of 'Age' column
sns.histplot(calories_data['Age'], kde=True)
#kde=Kernel Density Estimate (KDE) curve
```

Finding the correlation

1.Positive Correlation                
2.Negative Correlation

```python
correlation = calories_data.corr(numeric_only=True)
```

```python
#constructing a heatmap to understand the correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

Converting text data to numerical value

```python
calories_data['Gender'] = calories_data['Gender'].map({'male': 0, 'female': 1})
```

```python
calories_data.head()
```

Seperating Features and Target

```python
X=calories_data.drop(columns=['User_ID','Calories'],axis=1)
Y=calories_data['Calories']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training and test data

```python
X_train, X_test, Y_train, Y_test=train_test_split(X, Y, test_size=0.5, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

XGBoost Regressor

```python
#loading the model
model = XGBRegressor()
```

```python
#training the model with X_train
model.fit(X_train,Y_train)
```

Prediction on the Test data

```python
test_data_prediction=model.predict(X_test)
```

```python
print(test_data_prediction)
```

Mean Absolute Error

```python
mae=metrics.mean_absolute_error(Y_test, test_data_prediction)
```

```python
print("Mean Absolute Error = ",mae)
```
